## Ejercicio 1
$y'= -5y+5t^2+2t$
* $0 \le t \le 1$
* $y(t_0)=1/3$

## Comparación con la solución real
La solución real de la ecuación diferencial es:
$y(t)=t^2 +1/3e^{-5t}$

* Grafique la solución real y la aproximación obtenida con el método de Euler en una misma gráfica.



# Errores
Calcule el eeror relativo para diferentes valores de N.

```python
import math

# Definición de la función diferencial
def f(t: float, y: float) -> float:
    return (-5)*y + 5*(t**2) + 2*t

# Definición de parámetros del problema
a = 0
b = 1
y_t0 = 1/3
N = 10

# Resolución del problema mediante el método de Euler
y, t, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Cálculo de la solución exacta
y_exact = [(ti ** 2) + (1/3) * math.exp(-5 * ti) for ti in t]
error = [yi - y_exact_i for yi, y_exact_i in zip(y, y_exact)]

# Impresión de resultados
print("t:", t)
print("y:", y)
# Cálculo del error relativo
error_relativo = [abs(e / y_exact_i) for e, y_exact_i in zip(error, y_exact)]

# Impresión del error relativo
print("Error relativo:", error_relativo)
# Eliminación del arreglo error_relativo


# Cálculo del promedio del error relativo
promedio_error_relativo = sum(error_relativo) / len(error_relativo)
print("Promedio del error relativo:", promedio_error_relativo)
# Impresión del promedio del error relativo en notación científica
print("Promedio del error relativo (notación científica): {:.2e}".format(promedio_error_relativo))
```

```python
import matplotlib.pyplot as plt

# Resolución del problema mediante el método de Euler
y_aprox, t, h = ODE_euler(a=a, b=b, f=f, y_t0=y_t0, N=N)

# Cálculo de la solución exacta
y_real = [(ti ** 2) + (1/3) * math.exp(-5 * ti) for ti in t]

plt.figure(figsize=(8, 5))
plt.plot(t, y_real, label="Y_real (Exacta)", linestyle="dashed", color="blue")
plt.plot(t, y_aprox, label="Y_aprox (Euler)", marker="o", color="red")
plt.xlabel("Tiempo t")
plt.ylabel("Y(t)")
plt.title("Comparación entre Y_real y Y_aprox")
plt.legend()
plt.grid(True)
plt.show()
```
